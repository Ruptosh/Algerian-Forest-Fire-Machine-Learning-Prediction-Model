# Data Cleaning for Better Model Prediction 

Converting Object Dtype to Float Dtype for Model Accuracy in X_train dataset

```python
df = df.drop(columns = ('DC'), axis = 1)
df = df.drop(columns = ('FWI'), axis = 1) 
```

```python
df
```

```python
df.info()
```

```python
## Independent and Dependent Feature 

x = df.iloc[:, 1:-2]
y = df.iloc [ : , 0] 
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# Feature Scalling the Dataset

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
```

```python
scaler
```

```python
X_train = scaler.fit_transform(X_train) 
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Linear Regression Model Prediction 

```python
from sklearn.linear_model import LinearRegression 
regression = LinearRegression()  
```

```python
regression
```

```python
regression.fit(X_train, y_train) 
```

```python
# print the Coefficients 
print(regression.coef_)
```

```python
## print the intercept 
print(regression.intercept_)
```

```python
## Prediction for the Test data 
reg_pred = regression.predict(X_test)
```

```python
# Relationship Between Real Data & Predicted Data 
plt.scatter(y_test,reg_pred) 
plt.xlabel("Test Truth Data") 
plt.ylabel("Test Predicted Data")
```

```python
# Calculating Residual 
residuals = y_test - reg_pred
```

```python
residuals
```

```python
# Distribution of residual are Approximately Normal Distribution 
sns.displot(residuals) 
```

```python
sns.displot(residuals,kind ="kde")
```

```python
#Scatter plot
plt.scatter(reg_pred, residuals) 
```

```python
## Performance Metrics 
from sklearn.metrics import mean_squared_error 
from sklearn.metrics import mean_absolute_error 
print(mean_squared_error(y_test, reg_pred)) 
print(mean_absolute_error(y_test,reg_pred)) 
print(np.sqrt(mean_squared_error(y_test, reg_pred))) 
```

```python
#R Squared
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred) 
print (score) 
```

```python
## Adjusted R square 
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
